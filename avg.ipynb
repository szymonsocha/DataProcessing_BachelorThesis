```python
import pandas as pd
from datetime import datetime
import statistics
```

```python
kraje = ['global', 'us', 'gb', 'ae', 'ar', 'at', 'au', 'be', 'bg', 'bo', 'br', 'ca', 'ch', 'cl',
         'co', 'cr', 'cy', 'cz', 'de', 'dk', 'do', 'ec', 'ee', 'eg', 'es', 'fi', 'fr', 'gr',
         'gt', 'hk', 'hn', 'hu', 'id', 'ie', 'il', 'in', 'is', 'it', 'jp', 'kr', 'lt', 'lu',
         'lv', 'ma', 'mx', 'my', 'ni', 'nl', 'no', 'nz', 'pa', 'pe', 'ph', 'pl', 'pt', 'py',
         'ro', 'ru', 'sa', 'se', 'sg', 'sk', 'sv', 'th', 'tr', 'tw', 'ua', 'uy', 'vn', 'za']
```

```python
daty = pd.date_range(start="2020-01-03",end="2021-03-15")
```

```python
daty = daty.strftime('%Y-%m-%d')
```

```python
x = 0
```

```python
# Zakres dat dla ktorych wyliczona zostala mediana w Google Mobility Report
dolne_ogr = datetime(2020, 1, 3)
dolne_ogr = dolne_ogr.strftime('%Y-%m-%d')
gorne_ogr = datetime(2020, 2, 6)
gorne_ogr = gorne_ogr.strftime('%Y-%m-%d')
```

```python
for kraj in kraje: 
    df = pd.DataFrame(columns=['data', 'srednia'])
    odniesienie = pd.DataFrame(columns=['data', 'srednia'])
    for data in daty:
        try:
            dane = pd.read_csv('C:/Users/szymo/Desktop/Nauka/Licencjat/Dane/licencjat dane/daily/regional-'+kraj+'-daily-'+data+'.csv')
            dane = dane[1:]
            dane.columns = ['Position','Track Name','Artist','Streams','URL']
            dane = dane.set_index('Position')
            dane["Streams"] = pd.to_numeric(dane["Streams"])
            srednia = dane["Streams"].mean()
            df = df.append({'data': data, 'srednia' : srednia}, ignore_index=True)
            if dolne_ogr <= data <= gorne_ogr:
                odniesienie = odniesienie.append({'data': data, 'srednia' : srednia}, ignore_index=True)
                mediana = statistics.median(odniesienie["srednia"])
        except:
            try:
                dane = pd.read_csv('C:/Users/szymo/Desktop/Nauka/Licencjat/Dane/licencjat dane/daily/regional-'+kraj+'-daily-'+data+'-automat.csv',encoding='latin1')
                dane.columns = ['Position','Track Name','Artist','Streams','URL']
                dane = dane.set_index('Position')
                dane["Streams"] = pd.to_numeric(dane["Streams"])
                srednia = dane["Streams"].mean()
                df = df.append({'data': data, 'srednia' : srednia}, ignore_index=True)
                if dolne_ogr <= data <= gorne_ogr:
                    odniesienie = odniesienie.append({'data': data, 'srednia' : srednia}, ignore_index=True)
                    mediana = statistics.median(odniesienie["srednia"])
            except:
                print("|ERROR|  ", "kraj: ", kraj, "  data: ", data)
                x = x+1

    try:
        mediana = statistics.median(odniesienie["srednia"])
        iter = 0
        for data in df['data']:
            if data > gorne_ogr:
                temp = df.loc[df.data > gorne_ogr, 'srednia'].values[iter]
                if temp > mediana:
                    zmiana = ((temp/mediana)-1)*100
                    df.loc[df.data == data, 'zmiana'] = round(zmiana)
                elif temp < mediana:
                    zmiana = (-(1-(temp/mediana)))*100
                    df.loc[df.data == data, 'zmiana'] = round(zmiana)
                else:
                    zmiana = 0
                    df.loc[df.data == data, 'zmiana'] = round(zmiana)
                iter = iter + 1
    except:
        print("|PROBLEM Z MEDIANA| ", "kraj: ", kraj)
    df.to_csv('C:/Users/szymo/Desktop/daily_avg/'+kraj+'-daily_avg.csv', index=False)
print('')
print('Liczba bledow:', x)
print('Mediana: ', mediana)
```
